```python
 %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("./city_data.csv")
df1.head()
```

```python
df2 = pd.read_csv("./ride_data.csv")
df2.head()
```

```python
data_complete = pd.merge(df2, df1, how="right", on=["city"])
```

```python
data_complete.head()
```

```python
data_city = data_complete.groupby(['city'])
data_city.head()
```

```python
city_unique = data_complete.drop_duplicates('city') 
city_unique
```

```python
city_unique_grp = city_unique.groupby(['type'])
```

```python
rural_drivers = city_unique_grp['driver_count'].sum()
rural_drivers
```

```python
average_city = data_city['fare'].mean()
average_city
```

```python
city_drivers = city_unique.iloc[:,4]
```

```python
data_type = data_complete.groupby(['type'])
total_fare = data_complete['fare'].sum()
total_fare
```

```python
total_rides = data_city['fare'].count()
total_rides
```

```python
rural_fare = data_type['fare'].sum()
rural_fare
```

```python

colors = ["yellow", "lightskyblue", "lightcoral" ]
x_axis = total_rides
data = average_city
plt.scatter(x_axis, data, marker="o", facecolors= "red", edgecolors="black",
            s=city_drivers, alpha=0.75)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per city)")
plt.ylabel("Average Fair ($)")


```

```python
labels = "Rural", "Suburban", "Urban"
sizes = rural_fare
colors = ["yellow", "lightskyblue", "lightcoral" ]
explode = (0, 0, 0.1)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=120)
plt.title("% of Total Fares by City Type")
```
